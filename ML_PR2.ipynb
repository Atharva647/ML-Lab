```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6371

    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2

    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    distance = R * c

    return distance

```

```python
df['distance_km'] = haversine(df['pickup_latitude'], df['pickup_longitude'],
                              df['dropoff_latitude'], df['dropoff_longitude'])
```

```python
def remove_outlier(df,col):
    q1=df[col].quantile(0.25)
    q2=df[col].quantile(0.50)
    q3=df[col].quantile(0.75)
    
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_boubd=q1+1.5*iqr
    df[col]=np.clip(df[col],lower_bound,upper_boubd)
    
def treat_out(df,collist):
    for col in collist:
        remove_outlier(df,col)
```

```python
treat_out(df,df.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
```

```python
x=df.drop(['fare_amount'],axis=1)
```

```python
y=df['fare_amount']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=2)

LR = LinearRegression()

LR.fit(x_train, y_train)

y_pred_lr = LR.predict(x_test)
```

```python
LR.score(x_test,y_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(f"mae: {mean_absolute_error(y_test, y_pred_lr)}")
print(f"r2_score: {r2_score(y_test,y_pred_lr)}")
print(f"mse: {mean_squared_error(y_test, y_pred_lr)}")
```

```python
LE = Lasso(alpha=0.01, max_iter=1000)

LE.fit(x_train, y_train)

y_pred_lasso = LE.predict(x_test)
```

```python
print(f"mae: {mean_absolute_error(y_test, y_pred_lasso)}")
print(f"r2_score: {r2_score(y_test,y_pred_lasso)}")
print(f"mse: {mean_squared_error(y_test, y_pred_lasso)}")
```

```python
RE = Ridge(alpha=1.0)

RE.fit(x_train, y_train)

y_pred_ridge= RE.predict(x_test)
```

```python
print(f"mae: {mean_absolute_error(y_test, y_pred_ridge)}")
print(f"r2_score: {r2_score(y_test,y_pred_ridge)}")
print(f"mse: {mean_squared_error(y_test, y_pred_ridge)}")
```

```python
LR_acc=round(r2_score(y_test,y_pred_lr)*100,4)
Lasso_acc=round(r2_score(y_test,y_pred_lasso)*100,4)
Ridge_acc=round(r2_score(y_test,y_pred_ridge)*100,4)
```

```python
dic={'Model_Name':['Linear Regression','Lasso Regression','Ridge Regression'],'Accuracy':[LR_acc,Lasso_acc,Ridge_acc]}
```
